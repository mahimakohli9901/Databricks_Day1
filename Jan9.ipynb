```python
data = [
    (1, "Mahima", "IT", 60000, "Mumbai"),
    (2, "Amit", "HR", 45000, "Delhi"),
    (3, "Neha", "IT", 70000, "Bangalore"),
    (4, "Rahul", "Finance", 55000, "Mumbai"),
    (5, "Pooja", "HR", 48000, "Delhi")
]

columns = ["emp_id", "name", "department", "salary", "city"]
```

```python
df1 = spark.createDataFrame(data, columns)
display(df1)
```

```python
df1.filter(df1.department=="IT").show()
```

```python
from pyspark.sql.functions import avg

df1.groupBy("department") \
      .agg(avg("salary").alias("avg_salary")) \
      .show()
```
