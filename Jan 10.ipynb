```python
data = [
    (1001, "Electronics", "Mobile", 25000, "Mumbai"),
    (1002, "Electronics", "Laptop", 65000, "Delhi"),
    (1003, "Fashion", "Shoes", 3000, "Bangalore"),
    (1004, "Home", "Mixer", 4500, "Mumbai"),
    (1005, "Fashion", "T-Shirt", 1500, "Delhi"),
    (1006, "Electronics", "Headphones", 2000, "Pune"),
    (1007, "Home", "Vacuum Cleaner", 12000, "Bangalore")
]
```

```python
columns = ["order_id", "category", "product", "price", "city"]
```

```python
df = spark.createDataFrame(data, columns)
```

```python
df.show()
```

```python
df.printSchema()
```

```python
df.select("category","product","price").show()
#select few columns
```

```python
#select with alias
from pyspark.sql.functions import col,sum, avg, count
df.select(col("product").alias("product_name"),col("price")).show()

```

```python
df.filter((df.product=="Mobile")&(df.price>5000)).show()
#filter with multiple conditions
```

```python
df.groupBy("category").agg(sum("price").alias("Total_PRice")).show()
#group by and aggregate
#df.groupBy("category").agg(sum("price")
```

```python
df.orderBy(df.price.desc()).show()
```
