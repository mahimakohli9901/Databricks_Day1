```python
incremental_data = [
    ("CA-2026-99999", "Technology", 1200, 300),
    ("CA-2026-12345", "Furniture", 800, 150)
]

cols = ["Order ID", "Category", "Sales", "Profit"]

inc_df = spark.createDataFrame(incremental_data, cols)
inc_df.createOrReplaceTempView("incremental_updates")
```

```python
%sql
MERGE INTO superstore_delta t
USING incremental_updates s
ON t.`Order ID` = s.`Order ID`
WHEN MATCHED THEN
  UPDATE SET
    t.Sales = s.Sales,
    t.Profit = s.Profit
WHEN NOT MATCHED THEN
  INSERT *
```
